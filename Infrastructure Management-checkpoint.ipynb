```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 1000

# Material options
material_types = ['Ductile', 'GI', 'PVC', 'AC', 'HDPE']

# Generate the data
data = {
    'Asset': ['Intake Pipe'] * num_samples,  # Fixed value for asset type
    'Breaks in 10 Year': np.random.poisson(2, num_samples),  # Poisson distribution for breaks
    'Dia (mm)': np.random.choice(np.arange(50, 101, 10), size=num_samples),  # Diameters in increments of 10
    'Age': np.random.randint(30, 47, size=num_samples),  # Uniformly distributed ages
}

# Initialize material columns with zeros
for material in material_types:
    data[material] = [0] * num_samples

# Initialize other categorical variables with zeros
data['Sand'] = [0] * num_samples
data['Clay'] = [0] * num_samples
data['Trees'] = [0] * num_samples
data['No Trees'] = [0] * num_samples
data['Shallow Under 6'] = [0] * num_samples
data['Deep Bury'] = [0] * num_samples

# Assign one material, one soil type, one tree status, and one bury depth to each sample
for i in range(num_samples):
    chosen_material = np.random.choice(material_types)
    data[chosen_material][i] = 1
    
    soil_type = np.random.choice(['Sand', 'Clay'])
    data[soil_type][i] = 1
    
    tree_status = np.random.choice(['Trees', 'No Trees'])
    data[tree_status][i] = 1
    
    bury_depth = np.random.choice(['Shallow Under 6', 'Deep Bury'])
    data[bury_depth][i] = 1

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate Current Condition
def determine_condition(row):
    if row['Breaks in 10 Year'] < 2:
        condition = 1  # Good
    elif row['Breaks in 10 Year'] > 2:
        condition = 0  # Bad
    else:
        condition = 2  # Fair

    # Adjust condition based on age
    if row['Age'] > 40:
        condition = max(0, condition - 1)  # Age > 40 likely worse condition
    
    return condition

# Apply the function to each row
df['Current Condition'] = df.apply(determine_condition, axis=1)

# Save the DataFrame to an Excel file
df.to_excel('Simulated_Data.xlsx', index=False)
print("Dataset created and saved successfully.")

# Display first few rows of the DataFrame
print(df.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_df = pd.read_excel('Simulated_Data.xlsx')
column_df = data_df.columns
data_df = data_df[column_df[1:]]

# Prepare features and target variable
X = data_df.drop('Current Condition', axis=1)
y = data_df['Current Condition']

# Feature Engineering: Add polynomial features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define the model with regularization (L2 by default)
model = LogisticRegression(max_iter=10000, solver='liblinear')

# Define hyperparameters to tune
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
}

# Apply Grid Search
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best model
best_model = grid_search.best_estimator_

# Predict on test data
y_pred = best_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Best Parameters:", grid_search.best_params_)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Condition')
plt.ylabel('Actual Condition')
plt.title('Confusion Matrix')
plt.show()

# Perform cross-validation
cv_scores = cross_val_score(best_model, X_poly, y, cv=5, scoring='accuracy')
print("Cross-validation scores:", cv_scores)
print("Mean cross-validation score:", np.mean(cv_scores))

# Save the confusion matrix plot
plt.savefig('Improved_Confusion_Matrix.png')

# Save the classification report
with open('Improved_Classification_Report.txt', 'w') as f:
    f.write(class_report)
```

```python
def minChanges(n: int, k: int) -> int:
    n_bit = []
    k_bit = []
    while n > 1:
        if n % 2 == 0:
            n_bit.append(0)
            val = n / 2
            n = n - val
        else:
            n_bit.append(1)
            val = n // 2
            n = n -val
    else:
        n_bit.append(0)

    while k > 1:
        if k % 2 == 0:
            k_bit.append(0)
            val = k / 2
            k = k - val
        else:
            k_bit.append(1)
            val = k // 2
            k = k -val
    else:
        k_bit.append(0)
        
    if len(n_bit) > len(k_bit):
        diff = len(n_bit) - len(k_bit)
        for i in range(diff):
            k_bit.append(0)
    elif len(n_bit) < len(k_bit):
        diff = len(k_bit) - len(n_bit)
        for i in range(diff):
            n_bit.append(0)
    n_bit = list(reversed(n_bit))
    k_bit = list(reversed(k_bit))

    count = 0
    for i in range(len(n_bit)):
        if n_bit[i] != k_bit[i]:
            count += 1

    return count
```
